```python
import pandas as pd
books = pd.read_csv('books_cleaned.csv')
```

```python
books["categories"].value_counts().reset_index()
```

```python
category_mapping = {'Fiction' : "Fiction",
'Juvenile Fiction': "Children's Fiction",
'Biography & Autobiography': "Nonfiction",
'History': "Nonfiction",
'Literary Criticism': "Nonfiction",
'Philosophy': "Nonfiction",
'Religion': "Nonfiction",
'Comics & Graphic Novels': "Fiction",
'Drama': "Fiction",
'Juvenile Nonfiction': "Children's Nonfiction",
'Science': "Nonfiction",
'Poetry': "Fiction"}
```

```python
books["simplified_category"] = books["categories"].map(category_mapping)
```

```python
books
```

```python
books[books["simplified_category"].notna()].value_counts().reset_index()
```

```python
from transformers import pipeline

fiction_categories = ["Fiction" , "Nonfiction"]

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
sequence = books.loc[books["simplified_category"] == "Fiction", "description"].reset_index(drop=True)[0]
classifier(sequence, fiction_categories)
```

```python
import numpy as np
max_index = np.argmax(classifier(sequence, fiction_categories)["scores"])
max_label = classifier(sequence, fiction_categories)["labels"][max_index]
max_label
```

```python
def classify_book_description(description,categories):
    result = classifier(description, categories)
    max_index = np.argmax(result["scores"])
    return result["labels"][max_index]
```

```python
from tqdm import tqdm
actual_categories = []
predicted_categories = []
for index in tqdm(range(0, 300)):
    description = books.loc[books["simplified_category"] == "Fiction", "description"].reset_index(drop=True)[index]
    predicted_categories.append(classify_book_description(description, fiction_categories))
    actual_categories.append("Fiction")
```

```python
for index in tqdm(range(0, 300)):
    description = books.loc[books["simplified_category"] == "Nonfiction", "description"].reset_index(drop=True)[index]
    predicted_categories.append(classify_book_description(description, fiction_categories))
    actual_categories.append("Nonfiction")
```

```python
predictions_df = pd.DataFrame({
    "actual_category": actual_categories,
    "predicted_category": predicted_categories
})
```

```python
predictions_df.head()
```

```python
predictions_df["correct"] = predictions_df["actual_category"] == predictions_df["predicted_category"]
predictions_df["correct"] = predictions_df["correct"].map({True: 1, False: 0})
```

```python
predictions_df.head()
```

```python
predictions_df["correct"].sum() / predictions_df.shape[0]
```

```python
isbns = []
predicted_categories = []

missing_cats = books.loc[books["simplified_category"].isna(), ["isbn13","description"]]
missing_cats.head()
```

```python
for i in tqdm(range(0, missing_cats.shape[0])):
    isbn = missing_cats.iloc[i]["isbn13"]
    description = missing_cats.iloc[i]["description"]
    predicted_category = classify_book_description(description, fiction_categories)
    isbns.append(isbn)
    predicted_categories.append(predicted_category)
```

```python
missing_predictions_df = pd.DataFrame({"isbn13": isbns, "predicted_category": predicted_categories})
```

```python
missing_predictions_df.head()
```

```python
books = pd.merge(books, missing_predictions_df, on="isbn13", how="left")
books["simplified_category"] = books["simplified_category"].fillna(books["predicted_category"])
books = books.drop(columns=["predicted_category"])
```
