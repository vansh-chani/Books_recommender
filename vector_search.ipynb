```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import pandas as pd
books = pd.read_csv("books_cleaned.csv")
books.head()
```

```python
books["tag_description"].to_csv("books_tag_description.txt",sep = "\n", header = False,  index=False)
```

```python
raw_docs = TextLoader("books_tag_description.txt").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0,separator="\n")
documents = text_splitter.split_documents(raw_docs)
```

```python
documents[0]
```

```python
embedding_model = HuggingFaceEmbeddings(model_name="intfloat/e5-small-v2")
```

```python
db_books = Chroma.from_documents(
    documents,
    embedding=embedding_model,
)
```

```python
query = "A book to teach children about the nature"
results = db_books.similarity_search(query, k=10)
results
```

```python
query = "A detective mystery book by agatha christie"
results = db_books.similarity_search(query, k=10)
results
```

```python
books[books["isbn13"] == int(results[0].page_content.split(" ")[0])]
```

```python
def get_semantic_recommendations(query, k=10):
    results = db_books.similarity_search(query, k=k)
    res_isbn = [int(result.page_content.strip('"').split(" ")[0]) for result in results]
    pd_results = books[books["isbn13"].isin(res_isbn)]
    return pd_results
```

```python
get_semantic_recommendations("a book about wizards by j.k. rowling", k=10)
```
